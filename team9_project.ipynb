```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


plt.figure(figsize=(7,4), dpi=150)
df=pd.read_csv('uni.csv')
x=df['universities']
y=df['tuition_and_fees_for_2020']
plt.bar(x, y, color='c')
plt.xticks(rotation=90, size=7)
plt.xlabel('Universities',color='m', size=14)
plt.ylabel('tuition and fees (usd)',color='m', size=16)
plt.suptitle('Most Expensive Universisties in 2020',size=18, color='m')
plt.title('\n\nAll the informations was taken from armacad.com', size=9)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
plt.style.use('fivethirtyeight')
plt.figure(figsize=(8,4), dpi=150)
d=pd.read_csv('uni.csv')
x=d['universities']
w=0.2
b=np.arange(len(x))
b1=[i+w for i in b]
b2=[i-w for i in b]
y=d['qs_rankings_2021']
plt.bar(b,y, w, color='k',label='QS rank')

z=d['Place in US news 2021 ranking']
plt.bar(b2,z, w, color='yellow',label='US news rank')

a=d['place_in_times_higher_ed_2021_ranking']
plt.bar(b1,a, w,color='c',label='Times higher ed 2021 rank')
plt.xticks(b, x, rotation=90, size=9, color='k')

plt.xlabel('Universities', color='m', size=16)
plt.ylabel('Rankings', color='m', size=16)
plt.suptitle('    Top ten universities rankings according to three ranking systems ', color='m')
plt.title('All the informations was taken from armacad.com', size=9)
plt.legend(loc='upper right')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np

plt.style.use('seaborn')
plt.figure(figsize=(8,4), dpi=150)
d=pd.read_csv('usnews.csv')
x=d['universities']
w=0.2
b=np.arange(len(x))
b1=[i+w/2 for i in b]

y=d['tuition_and_fees']
z=d['median_starting_salary_of_alumni']

plt.bar(b-w/2,y, w, color='green',label='Tuition and fees')
plt.bar(b1,z, w, color='orange',label='median starting salary of alumni')

plt.xticks(b, x, rotation=90, size=9, color='k')

plt.xlabel('Universities', color='m', size=16)
plt.ylabel('in (USD)', color='m', size=16)
plt.suptitle('    Difference between tuition and starting salary of alumni ', color='m')
plt.title('All the informations was taken from usnews.com', size=9)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np

plt.style.use('seaborn')
plt.figure(figsize=(8,4), dpi=150)
d=pd.read_csv('usnews.csv')
x=d['universities']
w=0.2
b=np.arange(len(x))
b1=[i+w/2 for i in b]

y=d['national_students']
z=d['international_students']

plt.bar(b-w/2,y, w, color='green',label='national students')
plt.bar(b1,z, w, color='orange',label='International students')

plt.xticks(b, x, rotation=90, size=9, color='k')

plt.xlabel('Universities', color='m', size=16)
plt.ylabel('Students number', color='m', size=16)
plt.suptitle('    Difference between number of national and international students ', color='m')
plt.title('All the informations was taken from usnews.com', size=9)
plt.legend()

plt.show()
```
